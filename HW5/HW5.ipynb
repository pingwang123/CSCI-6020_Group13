```python
import numpy as np
import pandas as pd
 

print( init_data.keys())
print (init_data.keys())

data = pd.DataFrame(init_data.data, columns=[init_data.feature_names])
print (data.describe())

data = data.assign(target=pd.Series(init_data.target))
print (data.describe())
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
#check the correlations
data.corr()
```

```python
#check the importance of each feature
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X,y)
feature_imp=pd.Series(clf.feature_importances_,index=init_data.feature_names).sort_values(ascending=False)
print(feature_imp)
```

```python
#select two features ('worst perimeter' and 'worst concave points')
X1=data[['worst perimeter','worst concave points']]
X1.shape
```

```python
#tune the parameters
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform



param_grid = {'n_estimators' : range(10,100)}

rfsearch=RandomizedSearchCV(estimator=clf, param_distributions=param_grid,cv=5)

rfsearch.fit(X1,y)
print("Accuracy:",rfsearch.best_score_)
print("Accuracy/number of features :",rfsearch.best_score_/2 )
print(rfsearch.best_params_)
```

```python
#check the importance of "worst concave points"
X2=X1[['worst perimeter']]
rfsearch.fit(X2,y)
print("Accuracy:",rfsearch.best_score_)
print(rfsearch.best_params_)
```
